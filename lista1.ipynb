```python
import re

def parse_polynomial(poly_str):
    # Regular expression pattern for matching polynomial terms
    term_pattern = re.compile(r'([+-]?)\s*(\d*)(x(?:\^(\d+))?)?')
    
    terms = {}
    for match in term_pattern.finditer(poly_str):
        sign, coefficient_str, x_part, exponent_str = match.groups()
        
        # Skip empty matches
        if not (coefficient_str or x_part):
            continue
        
        # Determine the coefficient
        coefficient = int(coefficient_str) if coefficient_str else 1
        if sign == '-':
            coefficient = -coefficient
        
        # Determine the exponent
        if x_part:
            exponent = int(exponent_str) if exponent_str else 1
        else:
            exponent = 0
        
        # Insert term into the dictionary
        terms[exponent] = terms.get(exponent, 0) + coefficient
    
    return terms

# Example usage
poly_str = "10x^2 + 2x - 1"
parsed_terms = parse_polynomial(poly_str)

print(parsed_terms)
```

```python
import collections
import random

Card = collections.namedtuple('Card', ['rank', 'suit'])

class FrenchDeck:
    ranks = [str(n) for n in range(2, 11)] + list('JQKA')
    suits = 'spades diamonds clubs hearts'.split()

    def __init__(self):
        self._cards = [Card(rank, suit) for suit in self.suits for rank in self.ranks]

    def __len__(self):
        return len(self._cards)
    
    def __getitem__(self, position):
        return self._cards[position]

    def shuffle(self):
        random.shuffle(self._cards)

myDeck = FrenchDeck()
print(myDeck[1])
myDeck.shuffle()
```
